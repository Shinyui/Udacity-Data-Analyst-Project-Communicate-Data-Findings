## Investigation Overview

> The goal of this inbestigation is to find out the factors that might affect loan APR and outcome

## Dataset Overview

> This data set contains 113,937 loans with 81 variables on each
loan, including loan amount, borrower rate (or interest rate),
current loan status, borrower income, and many others.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv("prosperLoanData.csv")

# select columns needed
df_extracted = df.iloc[:,[2, 3, 4, 5, 7, 15, 16, 18, 19, 21, 25, 26, 51, 63, 77, 80]]
```

## APR Comparison of Homeowners

> After the investigation we found out that one of the factor that will affect APR is whether the borrower is homeowner or not. It seems like those who are houseowners tends to have lower APR than those who aren't. The reason behind this phenomenon could be the landing club can take house as collateral, hence they won't need to take much risks than those who don't have house.

```python
sns.boxplot(x=df_extracted["IsBorrowerHomeowner"], y=df_extracted["BorrowerAPR"])
plt.xlabel("Is Homeowners?")
plt.ylabel("APR")
plt.title("APR Comparison of Homeowners");
```

## Loan Status Comparision of Homeowner

> On the other hand we also found out that whether the borrower is homeowner or not will also affect the loan status. From the loan status count plot below we can see that the number of loan's being declared defaulted, chargedoff, or pass due, the non-homeowner's loan seems to have a higher possibilities to become defaulted, chargedoff or pastdue. 

```python
# ordered list of loan status
loan_status_order = ['Completed', 'FinalPaymentInProgress', 'Current',
                     'Past Due (1-15 days)', 
                     'Past Due (16-30 days)',
                     'Past Due (31-60 days)',
                     'Past Due (61-90 days)',
                     'Past Due (91-120 days)',
                     'Past Due (>120 days)',
                     'Defaulted','Chargedoff', 'Cancelled']

# create ordered categorical variable, using list
df_extracted['LoanStatus'] = pd.Categorical(df_extracted['LoanStatus'],
                                            categories=loan_status_order,
                                            ordered = True)
plt.figure(figsize=(25, 5))
sns.countplot(df_extracted["LoanStatus"], hue=df_extracted["IsBorrowerHomeowner"])
plt.xlabel("Loan Status")
plt.title("Loan Status Comparison of Homeowner");
```

## Loan Status Comparision of Homeowner with Loan Amount
If we further dive into the phenomenon and add the loan amount to y-axis we'll found out that eventhough that the non-homeowners' loan tend to have a high possibilities to be declared default or chargedoff but the amount is relatively small as well. Eventhough the homeowners are less likely to have there loan being declared defalted or charged off but when they do the amount is often higher than non-homeowner.

```python
# ordered list of loan status
loan_status_order = ['Completed', 'FinalPaymentInProgress', 'Current',
                     'Past Due (1-15 days)', 
                     'Past Due (16-30 days)',
                     'Past Due (31-60 days)',
                     'Past Due (61-90 days)',
                     'Past Due (91-120 days)',
                     'Past Due (>120 days)',
                     'Defaulted','Chargedoff', 'Cancelled']

# create ordered categorical variable, using list
df_extracted['LoanStatus'] = pd.Categorical(df_extracted['LoanStatus'],
                                            categories=loan_status_order,
                                            ordered = True)
plt.figure(figsize=(25, 5))
sns.countplot(y=df_extracted["LoanStatus"], hue=df_extracted["IsBorrowerHomeowner"])
plt.xlabel("Number of the status given")
plt.ylabel("Loan status")
plt.title("Number of the Status given from 2005 ~ 2014");
```
